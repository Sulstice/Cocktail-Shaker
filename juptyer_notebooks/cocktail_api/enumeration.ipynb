```
# Welcome to the Cocktail Shaker Enumeration functionality! 
# With machine learning becoming increasingly popular I figured it would be best to include the 
# enumeration functionality. 

# After the combinations are produced the user has the option to do a enumeration.

# Cocktail Shaker Imports
# ------------------------
from cocktail_shaker import Cocktail
from cocktail_shaker import PeptideBuilder

# Generate a peptide with 2 slots
peptide_backbone = PeptideBuilder(2)
```

```
# Generate the combinations

cocktail = Cocktail(peptide_backbone,ligand_library = ['Br', 'I'])
combinations = cocktail.shake()
print (combinations)
```

```
# Setting the complexity of the enumeration refers to how many tries it could find another way to 
# represent a 1D string (RDKit)

enumerations = cocktail.enumerate(enumeration_complexity='low')
print (enumerations)
print ("\n")
print ("Length Of Library: " + str(len(enumerations)))
```

```
# Lets take it up a notch
enumerations = cocktail.enumerate(enumeration_complexity='high')
print (enumerations)
print ("\n")
print ("Length Of Library: " + str(len(enumerations)))
```
