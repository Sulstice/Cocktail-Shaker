```
# Welcome to the Cocktail Shaker Inclusion of amino acids! 
# The amino acids come as a list of strings and if you choose to pass in the parameter then 19 amino acids
# will appended into your ligand library

# Cocktail Shaker Imports
# ------------------------
from cocktail_shaker import Cocktail
from cocktail_shaker import PeptideBuilder

# Generate a peptide with 2 slots
peptide_backbone = PeptideBuilder(2)

```

```
cocktail = Cocktail(peptide_backbone, ligand_library=['Br', 'I'], include_amino_acids=True)
combinations = cocktail.shake()
print (combinations)
```
